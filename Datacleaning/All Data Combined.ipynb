```python
#write function to convert date columns to datetime objects and trim dates for month>6 (after end of June)
#returns a new dataframe with "Date" column elements as datetime objects, starting 1/1 until 6/30 for any year
def trimDataframes(sample_df):
    #copied sample_df to change
    output_df = sample_df.copy()
    
    #convert "Date" column values to datetime objects
    output_df["Date"] = pd.to_datetime(output_df["Date"])
    
    #set "Date" column as index for the output_df
    output_df = output_df.set_index("Date")
    
    #remove all unwanted dates out of range (after 6/30)
    output_df = output_df.loc[output_df.index.month <= 6]
    
    #sort the values by date
    output_df = output_df.sort_values(by="Date")
    
    return output_df
```

```python
#trim 2017 data 

#ozone
o3_2017_trimmed_df = trimDataframes(o3_2017_df)
o3_2017_trimmed_df

#pm2.5
pm25_2017_trimmed_df = trimDataframes(pm25_2017_df)
pm25_2017_trimmed_df
```

```python
#trim 2018 data 
to_merge = []

#ozone
o3_2018_trimmed_df = trimDataframes(o3_2018_df)
o3_2018_trimmed_df

#pm2.5
pm25_2018_trimmed_df = trimDataframes(pm25_2018_df)
pm25_2018_trimmed_df

to_merge.append((o3_2018_trimmed_df, pm25_2018_trimmed_df))
```

```python
#trim 2019 data 

#ozone
o3_2019_trimmed_df = trimDataframes(o3_2019_df)
o3_2019_trimmed_df

#pm2.5
pm25_2019_trimmed_df = trimDataframes(pm25_2019_df)
pm25_2019_trimmed_df

to_merge.append((o3_2019_trimmed_df, pm25_2019_trimmed_df))
```

```python
#trim 2020 data 

#ozone
o3_2020_trimmed_df = trimDataframes(o3_2020_df)
o3_2020_trimmed_df

#pm2.5
pm25_2020_trimmed_df = trimDataframes(pm25_2020_df)
pm25_2020_trimmed_df

to_merge.append((o3_2020_trimmed_df, pm25_2020_trimmed_df))
```

```python
#merge all dataframes in to_merge with the earliest year dataframe
all_years_o3_df = o3_2017_trimmed_df.copy()
all_years_pm25_df = pm25_2017_trimmed_df.copy()

for o3_df, pm25_df in to_merge:
    all_years_o3_df = all_years_o3_df.append(o3_df)
    all_years_pm25_df = all_years_pm25_df.append(pm25_df)
    
print(all_years_o3_df.head(10))
print(all_years_o3_df.tail(10))
all_years_pm25_df.head()
```
